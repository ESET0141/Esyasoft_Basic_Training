```python
df.rename(columns= {'Engine HP':'HP','Engine Cylinders' : 'Cylinders', 'Transmission Type':'Transmission_type', 'highway MPG': 'Highway_MPG', 'city mpg': 'City_mpg' } ,inplace=True)
```

```python
dup_df = df[df.duplicated()]
print(dup_df)
```

```python
no_dup = df.drop_duplicates()
```

```python
no_dup.count()
```

```python
print(df.isnull().sum())
```

```python
clean_df = no_dup.dropna()
```

```python
clean_df.shape
```

```python
print(clean_df.isnull().sum())
```

```python
num_col = clean_df.select_dtypes(include=['number'])

Q1 = num_col.quantile(.25)
Q3 = num_col.quantile(.75)

IQR = Q3 - Q1
print(IQR)
```
