```python
# for checking all the data types

ds.dtypes
```

```python
#dropping a perticular column
# for column axis =1 and for rows axis = 0
ds = ds.drop(['datasetName'], axis = 1)

ds.head(12)
```

```python
# For renaming...

ds = ds.rename(columns={"about": "About", "link":"Link", "categoryName": "Category_Name", "cloud": "Cloud", "vintage": "Vintage"})

ds.head(12)
```

```python
ds.info()
```

```python
ds.describe()
```

```python
ds.shape
```

```python
# for checking duplicate entries...

duplicated_rows = ds[ds.duplicated()]

print("Number of duplicate rows : ", duplicated_rows)
```

```python
ds.count()
```

```python
# direct method to remove duplicate

ds = ds.drop_duplicates()

ds.head(6)
```

```python
ds.shape
```

```python
# Dropping missing or null values

print(ds.isnull().sum())
```

```python
ds = ds.dropna()  #not feasible since it drops the whole row... 
ds.count()
```
